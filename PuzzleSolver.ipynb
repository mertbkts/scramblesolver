```python
import numpy as np
import random
```

```python
red_door = 1
red_window = -1
green_door = 2
green_window = -2
blue_door = 3
blue_window = -3
yellow_door = 4
yellow_window = -4
```

```python
# Assigning numbers according to their shapes and colors
card6 = np.array([red_door, yellow_door, blue_window, green_window])
card5 = np.array([blue_window, red_window, yellow_door, yellow_window ])
card3 = np.array([blue_door, green_window, blue_window, red_door])
card1 = np.array([yellow_window, red_window, blue_door, green_door])
card8 = np.array([green_door, yellow_door, green_window, red_door])
card4 = np.array([green_door, blue_door, yellow_window, red_door])
card2 = np.array([green_window, blue_window, yellow_window, red_door])
card9 = np.array([green_window, yellow_window, blue_door, red_door])
card7 = np.array([green_door, red_window, yellow_door, blue_window])
```

```python
# Assigning deck
deck = np.array([card1, card2, card3, 
                card4, card5, card6, 
                card7, card8, card9])

played_cards = np.zeros_like(deck)
```

```python
# Solver

counter = 0

def solver(previous_card, whole_deck, played, side):
    
    global counter

    if side == 'left':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if deck[previous_card][0] + np.roll(card, i)[2] == 0:
                    card_index = np.where((deck == card).all(axis=1))[0][0]

                    played[0] = deck[previous_card]
                    played[1] = np.roll(card, i)

                    whole_deck = np.delete(whole_deck, np.where((whole_deck == deck[previous_card]).all(axis=1)), 0)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'down') == True:
                        print ("Second card number is",card_index+1, ", turn clockwise", i, "times")
                        print("First card number is", previous_card+1) 
                        print("Number of compared cards: ", counter)
                        counter = 0
                        return
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        whole_deck = np.append(whole_deck, [deck[previous_card]], 0)    
                        continue
        else:
            solver(previous_card + 1, deck, played_cards, 'left')

    elif side == 'down':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if played[1][3] + np.roll(card, i)[1] == 0:
                   
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    played[2] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)
                    
                    if solver(card_index, whole_deck, played, 'down_right') == True:
                        print ("Third card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue
                    
    elif side == 'down_right':
        
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if (played[0][3] + np.roll(card, i)[1] == 0 and played[2][2] + np.roll(card, i)[0] == 0) :
                   
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    played[3] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'right') == True:
                        print ("Fourth card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue
    elif side == 'right':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if played[3][2] + np.roll(card, i)[0] == 0:
                   
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    played[4] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'up_right') == True:
                        print ("Fifth card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue 
    elif side == 'up_right':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if (played[4][1] + np.roll(card, i)[3] == 0 and
                    played[0][2] + np.roll(card, i)[0] == 0):
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    
                    played[5] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'up') == True:
                        print ("Sixth card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue 
                    
    elif side == 'up':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if played[5][1] + np.roll(card, i)[3] == 0:
                   
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    played[6] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'left_up') == True:
                        print ("Seventh card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue 
    elif side == 'left_up':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if (played[0][1] + np.roll(card, i)[3] == 0 and
                    played[6][0] + np.roll(card, i)[2] == 0):
                   
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    played[7] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    if solver(card_index, whole_deck, played, 'last') == True:
                        print ("Eighth card number is",card_index+1, ", turn clockwise", i, "times")
                        return True
                    else:
                        whole_deck = np.append(whole_deck, [card], 0)
                        continue 
    elif side == 'last':
        for card in whole_deck:
            for i in range(4):
                counter += 1
                if (played[7][0] + np.roll(card, i)[2] == 0 and
                    played[1][1] + np.roll(card, i)[3] == 0):
                
                    card_index = np.where((deck == card).all(axis=1))[0][0]
                    print ("Ninth card number is",card_index+1, ", turn clockwise", i, "times")
                    played[8] = np.roll(card, i)
                    whole_deck = np.delete(whole_deck, np.where((whole_deck == card).all(axis=1)), 0)

                    return True    
```
